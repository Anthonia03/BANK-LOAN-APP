```python
def compound_interest(principal, rate, periods_per_year, years):

    rate = rate / 100

    
    periods = periods_per_year * years
    amount = principal * (1 + rate / periods_per_year) ** periods
    interest = amount - principal
    return interest, amount

def main():
    print("Welcome to the Loan Calculator App")

   
    principal = float(input("Enter the loan amount: "))
    rate = float(input("Enter the annual interest rate (in percentage): "))
    periods_per_year = int(input("Enter the number of times you want to pay in a year: "))
    years = int(input("Enter the number of years you plan to pay through: "))

    
    total_interest, total_amount = compound_interest(principal, rate, periods_per_year, years)

    
    print("\nLoan details:")
    print("Principal amount: ${:.2f}".format(principal))
    print("Annual interest rate: {}%".format(rate))
    print("Number of times to pay per year:", periods_per_year)
    print("Number of years to pay through:", years)

    
    remaining_amount = total_amount

   
    print("\nPayment Progress:")
    while remaining_amount > 0:
        print("Remaining amount: ${:.2f}".format(remaining_amount))
        payment = float(input("Enter the amount you want to pay this time: "))
        if payment > remaining_amount:
            payment = remaining_amount
        remaining_amount -= payment
        print("Amount paid this time: ${:.2f}".format(payment))
        print("Remaining amount after payment: ${:.2f}".format(remaining_amount))
        print("-----------------------------")

    
    print("\nTotal interest paid: ${:.2f}".format(total_interest))

if __name__ == "__main__":
    main()
```

```python
class Loan:
    def __init__(self, principal, rate, periods_per_year, years):
        self.principal = principal
        self.rate = rate
        self.periods_per_year = periods_per_year
        self.years = years
        self.total_interest, self.total_amount = self.compound_interest()
        self.remaining_amount = self.total_amount

    def compound_interest(self):
        """
        Function to calculate compound interest
        """
        # Convert rate from percentage to decimal
        rate = self.rate / 100

        # Calculate compound interest
        periods = self.periods_per_year * self.years
        amount = self.principal * (1 + rate / self.periods_per_year) ** periods
        interest = amount - self.principal
        return interest, amount

    def make_payment(self, payment):
        """
        Function to make a payment towards the loan
        """
        if payment > self.remaining_amount:
            payment = self.remaining_amount
        self.remaining_amount -= payment
        return payment

    def get_remaining_amount(self):
        """
        Function to get the remaining amount of the loan
        """
        return self.remaining_amount


class LoanManager:
    def __init__(self):
        self.loans = []

    def add_loan(self, loan):
        self.loans.append(loan)

    def total_borrowed(self):
        return sum(loan.principal for loan in self.loans)

    def average_income_per_month(self):
        total_interest = sum(loan.total_interest for loan in self.loans)
        return total_interest / (len(self.loans) * loan.years * 12)

    def highest_earning_month(self):
        max_month = max(self.loans, key=lambda loan: loan.total_interest)
        return max_month.total_interest

    def predict_next_months(self, num_months):
        total_interest_per_month = sum(loan.total_interest / (loan.years * 12) for loan in self.loans)
        return total_interest_per_month * num_months

    def make_payment(self, index, payment):
        """
        Function to make a payment towards a specific loan
        """
        return self.loans[index].make_payment(payment)

    def get_remaining_amount(self, index):
        """
        Function to get the remaining amount of a specific loan
        """
        return self.loans[index].get_remaining_amount()


def main():
    loan_manager = LoanManager()

    while True:
        print("\n1. Add a loan")
        print("2. Make a payment")
        print("3. View company statistics")
        print("4. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            principal = float(input("Enter the loan amount: "))
            rate = float(input("Enter the annual interest rate (in percentage): "))
            periods_per_year = int(input("Enter the number of times you want to pay in a year: "))
            years = int(input("Enter the number of years you plan to pay through: "))
            loan = Loan(principal, rate, periods_per_year, years)
            loan_manager.add_loan(loan)
            print("Loan added successfully!")
        elif choice == '2':
            index = int(input("Enter the loan index to make a payment: "))
            payment = float(input("Enter the payment amount: "))
            remaining_amount = loan_manager.make_payment(index, payment)
            print("Remaining amount for loan {}: ${:.2f}".format(index, remaining_amount))
        elif choice == '3':
            print("\nCompany Statistics:")
            print("Total money borrowed out: ${:.2f}".format(loan_manager.total_borrowed()))
            print("Average income per month: ${:.2f}".format(loan_manager.average_income_per_month()))
            print("Highest earning month: ${:.2f}".format(loan_manager.highest_earning_month()))
            num_months = int(input("Enter the number of months to predict: "))
            print("Expected money over the next {} months: ${:.2f}".format(num_months, loan_manager.predict_next_months(num_months)))
        elif choice == '4':
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please try again.")


if __name__ == "__main__":
    main()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

class Loan:
    def __init__(self, principal, rate, periods_per_year, years):
        self.principal = principal
        self.rate = rate
        self.periods_per_year = periods_per_year
        self.years = years
        self.total_interest, self.total_amount = self.compound_interest()
        self.remaining_amount = self.total_amount

    def compound_interest(self):
        """
        Function to calculate compound interest
        """
        # Convert rate from percentage to decimal
        rate = self.rate / 100

        # Calculate compound interest
        periods = self.periods_per_year * self.years
        amount = self.principal * (1 + rate / self.periods_per_year) ** periods
        interest = amount - self.principal
        return interest, amount

    def make_payment(self, payment):
        """
        Function to make a payment towards the loan
        """
        if payment > self.remaining_amount:
            payment = self.remaining_amount
        self.remaining_amount -= payment
        return payment

    def get_remaining_amount(self):
        """
        Function to get the remaining amount of the loan
        """
        return self.remaining_amount


class LoanManager:
    def __init__(self):
        self.loans = []

    def add_loan(self, loan):
        self.loans.append(loan)

    def total_borrowed(self):
        return sum(loan.principal for loan in self.loans)

    def average_income_per_month(self):
        total_interest = sum(loan.total_interest for loan in self.loans)
        return total_interest / (len(self.loans) * loan.years * 12)

    def highest_earning_month(self):
        max_month = max(self.loans, key=lambda loan: loan.total_interest)
        return max_month.total_interest

    def predict_next_months(self, num_months):
        total_interest_per_month = sum(loan.total_interest / (loan.years * 12) for loan in self.loans)
        return total_interest_per_month * num_months

    def make_payment(self, index, payment):
        """
        Function to make a payment towards a specific loan
        """
        return self.loans[index].make_payment(payment)

    def get_remaining_amount(self, index):
        """
        Function to get the remaining amount of a specific loan
        """
        return self.loans[index].get_remaining_amount()

    def regression_analysis(self, income_data):
        """
        Function to perform regression analysis on income data
        """
        X = np.array(range(1, len(income_data) + 1)).reshape(-1, 1)
        y = np.array(income_data)
        model = LinearRegression().fit(X, y)
        return model.coef_[0], model.intercept_


def main():
    loan_manager = LoanManager()

    while True:
        print("\n1. Add a loan")
        print("2. Make a payment")
        print("3. View company statistics")
        print("4. Perform regression analysis")
        print("5. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            principal = float(input("Enter the loan amount: "))
            rate = float(input("Enter the annual interest rate (in percentage): "))
            periods_per_year = int(input("Enter the number of times you want to pay in a year: "))
            years = int(input("Enter the number of years you plan to pay through: "))
            loan = Loan(principal, rate, periods_per_year, years)
            loan_manager.add_loan(loan)
            print("Loan added successfully!")
        elif choice == '2':
            index = int(input("Enter the loan index to make a payment: "))
            payment = float(input("Enter the payment amount: "))
            remaining_amount = loan_manager.make_payment(index, payment)
            print("Remaining amount for loan {}: ${:.2f}".format(index, remaining_amount))
        elif choice == '3':
            print("\nCompany Statistics:")
            print("Total money borrowed out: ${:.2f}".format(loan_manager.total_borrowed()))
            print("Average income per month: ${:.2f}".format(loan_manager.average_income_per_month()))
            print("Highest earning month: ${:.2f}".format(loan_manager.highest_earning_month()))
            num_months = int(input("Enter the number of months to predict: "))
            print("Expected money over the next {} months: ${:.2f}".format(num_months, loan_manager.predict_next_months(num_months)))
        elif choice == '4':
            num_months = int(input("Enter the number of months for which you have income data: "))
            income_data = []
            for i in range(num_months):
                income = float(input("Enter income for month {}: ".format(i + 1)))
                income_data.append(income)
            slope, intercept = loan_manager.regression_analysis(income_data)
            print("Regression Analysis:")
            print("Slope: {:.2f}".format(slope))
            print("Intercept: {:.2f}".format(intercept))
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please try again.")


if __name__ == "__main__":
    main()
```
